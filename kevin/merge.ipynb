```python
jordan=pd.read_csv('../yazan/datasets/jordan_boringness_count.csv')
```

```python
jordan.columns=['boringness','songs_jordan']
```

```python
jordan.head()
```

```python
total_jordan=jordan['songs_jordan'].sum()
jordan['boringness_songs']=jordan['boringness']*jordan['songs_jordan']
total_boringness_songs_jordan=jordan['boringness_songs'].sum()
mean_jordan=round(total_boringness_songs_jordan/total_jordan,2)
mean_jordan
```
