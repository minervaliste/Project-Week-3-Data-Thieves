
# Country by genre (percentages)

### Spain

```python
genre_country_percentage = genre_country_minimal[['track_name', 'country_name', 'genre_name']].copy()
genre_percentage_Spain = genre_country_percentage[genre_country_percentage.country_name == 'Spain'].copy()
```

```python
genre_percentage_Spain = genre_country_percentage[genre_country_percentage.country_name == 'Spain'].copy()
```

```python
genre_percentage_Spain.drop(['country_name'], axis = 1, inplace = True)
sum_genres_spain = genre_percentage_Spain.groupby('genre_name', as_index = False).count()
sum_genres_spain.columns = ['genre_name', 'number_tracks']
sum_genres_spain['percentage'] = sum_genres_spain['number_tracks']/(sum_genres_spain['number_tracks'].sum())
sum_genres_spain.head(3)
```

```python
sum_genres_spain.to_csv('percentage_genre_spain.csv', sep = ';')
```
